The method most commonly used to get an overview of the data, describe() ,has a significant flaw.

It does not have any information on categorical data.

This new method ,however, does work for categorical data as well. It goes by the name of pandas_profiling.



## Dataset used for testing purpose

This dataset includes jobs put by NHS relating to doctors only between the period of March 2019 to October 2019.

[NHS jobs](https://www.kaggle.com/homelesssandwich/nhs-jobs)

```python
import pandas as pd
import numpy as np
import time
nhs = pd.read_csv("jobs_raw.csv")
#To display all the columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

nhs.head()
```

```python
start = time.time()
nhs.describe()
end = time.time()
# To record time taken for execution
dtime = end - start
nhs.describe()
```

Before I proceed with the crazy , new and revolutionary way of data analysis ( Again, sorry!) , something that caught my eye was the min salary being 103 GBP, which makes no sense for a doctor's job. 

```python
nhs[nhs["json_salary_min"] == nhs["json_salary_min"].min()]
```

Ah this is most likely a mistake where it is supposed to be 103,000 GBP. This can be confirmed from the json_salary_value column with value given as £76,761 - £103.

I wonder what the maximum salary is then.

```python
nhs[nhs["json_salary_max"] == nhs["json_salary_max"].max()]
```

Okay, someone definitely added a bit more zeroes than usual in the json_salary max column. 3 more zeroes to be precise.

Anyways , moving on...

While we have 29 columns in the dataset , describe gives us information for only 4 columns. This is especially worse if your dataset is riddled with issues as we saw above. 

Introducing the new method, *drumroll*

Pandas Profiling.

Here's how its used.



```python
!pip install pandas_profiling
import pandas_profiling
```

I am getting the requirement already satisfied message because I already installed it previously. You will have probably have to download the library.

Now that we have imported it, lets use it.

```python
start = time.time()
pandas_profiling.ProfileReport(nhs)
end = time.time()
# To record time taken for execution
ptime = end - start
pandas_profiling.ProfileReport(nhs)
```

And Voila! , right off the bat, we see that we have 36732 missing entries and 98 duplicate entries.

We also see a ton of very useful information regarding the categorical variables :

1.  department has 3631 (13.2%) missing values
    pay_band has 11642 (42.4%) missing values	
    working_pattern has 4442 (16.2%) missing values	

2.  staff_group has constant value "Medical & Dental" which shows that this column is quite useless and can be dropped.

We also see that a ton of other information regarding the columns including missing values in depth, correlations and so on.



Compared to describe() , this method is significantly better right?

For the most part yes. The information provided is much more and easily understandable compared to describe().

However , while this dataset is definitely not big (27482 rows) , it still took quite a lot of time to generate the report.

Let us compare the values.

```python
print ("Time taken for describe to execute : ",dtime)
print ("Time taken for pandas profiling to execute : ",ptime)
```

A difference of 30 seconds is definitely something that cannot be ignored. The time taken would further increase to larger datasets to a point where it is not feasible enough to use this method.

## Conclusion

While pandas profiling is excellent for smaller datasets , it is definitely not recommended for larger ones as the time taken is significantly high.
